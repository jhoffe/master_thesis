# Speaker CV

```python
import pandas as pd

# reports/metrics/lillelyd_finetune_average_metrics.parquet

# load from parquet
def load_from_parquet(parquet_file: str) -> pd.DataFrame:
    df = pd.read_parquet(parquet_file)
    return df

df = load_from_parquet("../reports/metrics/lillelyd_finetune_speaker_average_metrics.parquet")
```

```python
df.columns
```

```python
# filter for "parakeet" in model name
df_parakeet = df[df['model'].str.contains("parakeet")]

def extract_latex_rows(df: pd.DataFrame) -> str:
    latex_rows = ""
    for model in df['model'].unique():
        df_model = df[df['model'] == model]
        lillelyd_wer = 100*df_model[df_model['dataset_name'] == 'lillelyd']['WER'].values[0]
        lillelyd_pm_ci = 100*(df_model[df_model['dataset_name'] == 'lillelyd']['WER_ci_upper'].values[0] - lillelyd_wer/100)
        lillelyd_sem = df_model[df_model['dataset_name'] == 'lillelyd']['avg_semantic_distance'].values[0]
        coral_wer = 100*df_model[df_model['dataset_name'] == 'coral-v2']['WER'].values[0]
        coral_pm_ci = 100*(df_model[df_model['dataset_name'] == 'coral-v2']['WER_ci_upper'].values[0] - coral_wer/100)
        coral_sem = df_model[df_model['dataset_name'] == 'coral-v2']['avg_semantic_distance'].values[0]
        fleurs_wer = 100*df_model[df_model['dataset_name'] == 'fleurs']['WER'].values[0]
        fleurs_pm_ci = 100*(df_model[df_model['dataset_name'] == 'fleurs']['WER_ci_upper'].values[0] - fleurs_wer/100)
        fleurs_sem = df_model[df_model['dataset_name'] == 'fleurs']['avg_semantic_distance'].values[0]
        # the format is lillelyd WER pm CI & lillelyd Semdist & coral WER pm CI & coral Semdist & fleurs WER pm CI & fleurs Semdist \\
        print(f"Model - {model}: ${lillelyd_wer:.2f} \\pm {lillelyd_pm_ci:.2f}$ & ${lillelyd_sem:.3f}$ & ${coral_wer:.2f} \\pm {coral_pm_ci:.2f}$ & ${coral_sem:.3f}$ & ${fleurs_wer:.2f} \\pm {fleurs_pm_ci:.2f}$ & ${fleurs_sem:.3f}$ \\\\\n")
    return latex_rows
```

```python
parakeet_rows = extract_latex_rows(df_parakeet)
print(parakeet_rows)
```

```python
df_canary = df[df['model'].str.contains("canary")]

canary_rows = extract_latex_rows(df_canary)
```
