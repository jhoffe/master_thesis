```python
# load jsonl manifest from ../data/emotale/data/manifest.jsonl

import json
from pathlib import Path
from typing import List, Dict, Optional
import os
import pandas as pd

def load_manifest(manifest_path: str) -> List[Dict]:
    """Load a JSONL manifest file."""
    manifest = []
    with open(manifest_path, 'r') as f:
        for line in f:
            manifest.append(json.loads(line))
    return manifest

# make into df
def manifest_to_df(manifest: List[Dict]) -> pd.DataFrame:
    """Convert a list of dictionaries to a pandas DataFrame."""
    return pd.DataFrame(manifest)
# Example usage
manifest = load_manifest('../data/processed/LilleLyd/manifest.jsonl')

df = manifest_to_df(manifest)
```

```python
# sum the durations
total_duration = df['duration'].sum()
print(f"Total duration: {total_duration} seconds")
# in minutes
print(f"Total duration: {total_duration / 60} minutes")
# group by emotion and sum durations
duration_by_emotion = df.groupby('emotion')['duration'].sum()
print("Duration by emotion (minutes):")
print(duration_by_emotion/60)

# group by sentence and sum durations
duration_by_sentence = df.groupby('text')['duration'].sum()
print("Duration by sentence (minutes):")
print(duration_by_sentence/60)

# groupby participant_id and count number of (F)emale and (M)ale
gender_counts = df.groupby('participant_id')['gender'].first().value_counts()
print("Gender counts:")
print(gender_counts)
```
